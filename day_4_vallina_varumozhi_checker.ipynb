```python
import re

def check_first_letter_only():
    # ஒரு வாக்கியம்
    text = "அந்த தோட்டம்" # இதில் 'தோ' என்பது 'த்' + 'ஓ'

    # முறை 1: Lookahead (முன்னோக்கிப் பார்த்தல்) - இது எழுத்தை பிடிக்காது, செக் மட்டும் செய்யும்
    # இதில் \s+ என்பது இடைவெளி, (?=[கசதப]) என்பது அடுத்து வருவது வல்லினமா என்று பார்க்கும்
    pattern_check = r"\s+(?=[கசதப])"

    # முறை 2: நேரடித் தேடல் (அனைத்து உயிர்மெய்களையும் சேர்த்து)
    # க முதல் ப வரை உள்ள வரிசையில் தொடங்கும் எந்த உயிர்மெய் எழுத்தையும் இது பிடிக்கும்
    pattern_capture = r"\s+([க-ப][\u0BBE-\u0BCD]?)"

    match = re.search(pattern_capture, text)

    if match:
        # match.group(1) என்பது முதல் எழுத்தை மட்டும் தரும்
        print(f"வாக்கியம்: {text}")
        print(f"வருமொழியின் முதல் எழுத்து: {match.group(1)}")
        print("முடிவு: இது வல்லினத்தில் தொடங்குகிறது!")

check_first_letter_only()
```
