```python
import re

def dependency_logic_tester(sentence):
    # Mapping markers to their structural roles
    rules = {
        "ை$": "Direct Object (obj)",
        "(க்)?கு$": "Indirect Object/Recipient (iobj)",
        "ால்$": "Instrument (obl)",
        "இல்$": "Location (loc)"
    }

    tokens = sentence.split()
    structure = []

    for token in tokens:
        role = "Subject (nsubj) or Root" # Default
        for pattern, label in rules.items():
            if re.search(pattern, token):
                role = label
                break
        structure.append(f"{token} -> {role}")

    return structure

# Example: "கத்தியால் பழத்தை வெட்டினான்" (He cut the fruit with a knife)
test_sentence = "கத்தியால் பழத்தை வெட்டினான்"
print("\n".join(dependency_logic_tester(test_sentence)))
```
