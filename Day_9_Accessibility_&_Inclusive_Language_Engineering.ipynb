```python
import re

def accessibility_text_cleaner(raw_text, link_label="இந்த இணைப்பை சொடுக்கவும்"):
    # 1. Regex to find URLs
    url_pattern = r'https?://\S+'

    # 2. Replace URL with a clear Tamil label for Screen Readers
    # This prevents the AI from reading 100 random characters.
    clean_text = re.sub(url_pattern, link_label, raw_text)

    # 3. Handle Hashtags for Screen Readers (CamelCase check)
    # We ensure hashtags are recognizable as separate words
    clean_text = re.sub(r'#([a-z])', r'# \1', clean_text) # Adds a tiny pause

    return clean_text

# Example from your FB post
raw_post = "Join us! https://forms.office.com/Pages/ResponsePage.aspx?id=DQSI... #accessibility"
print("Before Engineering:", raw_post)
print("After Engineering:", accessibility_text_cleaner(raw_post))
```
