```python
import re

def regex_day_3_stemmer():
    # A list of Tamil plural words
    words = ["மலர்கள்", "வீடுகள்", "பூக்கள்", "பெண்கள்", "ஆசிரியர்கள்"]

    # THE FORMULA: (க்)?கள்$
    # (க்)? -> Matches the letter 'க்' zero or one time (optional).
    # கள்   -> Matches the literal characters 'கள்'.
    # $     -> The "End Anchor". This ensures we only match the end of the word.
    suffix_pattern = r"(க்)?கள்$"

    stemmed_words = []
    for word in words:
        # re.sub(pattern, replacement, string)
        # We replace the suffix with an empty string ""
        root = re.sub(suffix_pattern, "", word)
        stemmed_words.append(root)

    # Display results
    print(f"Plural Words: {words}")
    print(f"Root Words:   {stemmed_words}")

# Run the Day 3 experiment
regex_day_3_stemmer()
```
