```python
import re

def regex_day_2():
    # A mixed sentence with English and Tamil
    mixed_text = "I love Tamil (தமிழ்) and Python (பைதான்)."

    # THE FORMULA:
    # \u0B80-\u0BFF is the range of all Tamil characters.
    # [] means "match any one character in this range".
    # + means "match one or more characters in a row" (to get the whole word).
    tamil_pattern = r"[\u0B80-\u0BFF]+"

    matches = re.findall(tamil_pattern, mixed_text)

    print(f"Input: {mixed_text}")
    print(f"Tamil words found: {matches}")

# Execution
regex_day_2()
```
