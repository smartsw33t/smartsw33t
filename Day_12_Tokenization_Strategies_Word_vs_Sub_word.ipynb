```python
import re

def subword_tokenization_sim(text):
    # Common Tamil suffixes that carry meaning
    suffixes = r"(ங்கள்|களை|ஐ|க்கு|இலிருந்து|ஆல்)"

    words = text.split()
    tokens = []

    for word in words:
        # If the word ends with a common suffix, split it
        match = re.search(f"(.*){suffixes}$", word)
        if match:
            # We add '##' to show it's a sub-word unit (Standard NLP notation)
            tokens.append(match.group(1))
            tokens.append(f"##{match.group(2)}")
        else:
            tokens.append(word)

    return tokens

# Test the strategy
sentence = "மரங்கள் காற்றில் அசைந்தன"
print(f"Original: {sentence}")
print(f"Tokenized: {subword_tokenization_sim(sentence)}")
```
