```python
import re

def tamil_stopword_filter(text):
    # 1. List of common standalone stop-words
    stop_words = {"மற்றும்", "என்று", "இந்த", "அந்த", "ஒரு", "ஆகிய", "என"}

    # 2. Pattern for common 'noise' suffixes (like 'also', 'only', 'even')
    # -um (also), -than (only), -vathu (at least)
    suffix_noise = r"(உம்|தான்|ஆவது)$"

    words = text.split()
    filtered_words = []

    for word in words:
        # Step A: Check if the word is in the standalone list
        if word in stop_words:
            continue

        # Step B: Strip frequent noise suffixes to find the 'core' word
        # We use re.sub to remove the suffix if it exists
        core_word = re.sub(suffix_noise, "", word)

        filtered_words.append(core_word)

    return " ".join(filtered_words)

# Example
sample = "அவனும் மற்றும் இந்த மரமும் அழகானவை."
print(f"Original: {sample}")
print(f"Filtered: {tamil_stopword_filter(sample)}")
```
