```python
import re

def evaluate_ai_response(user_input, ai_output):
    report = {"Passed": [], "Failed": []}

    # Test 1: Did the AI recognize the name 'Achu'? (NER Check)
    if "Achu" in ai_output or "அச்சு" in ai_output:
        report["Passed"].append("Entity Recognition (Person)")
    else:
        report["Failed"].append("Entity Recognition (Person)")

    # Test 2: Did the AI handle the complex Tamil word? (Morphology Check)
    if "வாழைப்பழம்" in ai_output:
         report["Passed"].append("Complex Script Integrity")
    else:
         report["Failed"].append("Complex Script Integrity")

    return report

# Simulated Test
input_prompt = "Achu வாழைப்பளம் சாப்பிட்டாள்." # Note the intentional typo 'ளம்' instead of 'ழம்'
ai_resp = "Achu banana saapitral." # AI simplified it too much

print(f"LQA Report: {evaluate_ai_response(input_prompt, ai_resp)}")
```
