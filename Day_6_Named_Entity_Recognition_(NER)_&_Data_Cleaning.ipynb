```python
import re

def day6_ner_processor(text):
    # Dictionary of known locations (Gazetteer)
    locations = ["மதுரை", "சென்னை", "கோவை", "Madurai", "Chennai"]

    # Regex for Person markers (Honorifics)
    # This looks for 'Thiru' or 'Avargal' or 'Shri'
    person_marker_pattern = r"(திரு\.\s?|திருமதி\.\s?|Shri\.\s?)([அ-ஹ]+)"

    entities = {"Persons": [], "Locations": []}

    # 1. Check for Locations using our list
    for loc in locations:
        if loc in text:
            entities["Locations"].append(loc)

    # 2. Check for Persons using Regex markers
    person_match = re.search(person_marker_pattern, text)
    if person_match:
        entities["Persons"].append(person_match.group(2))

    return entities

# Test Input
sample_text = "திரு. ஜெயசீலன் சென்னை சென்றார்."
print(f"Entities Found: {day6_ner_processor(sample_text)}")
```
