```python
import re

def process_mixed_content():
    # 5-line mixed content (Tamil, Tanglish, and English)
    mixed_data = [
        "இயற்கை மிகவும் அழகானது.",                 # Native Tamil
        "The weather is very pleasant today.",       # Pure English
        "Innaiku climate romba nalla irukku!",       # Tanglish
        "Next week nammu ooru polama?",              # Tanglish + English mix
        "கணினி மொழியியல் என்பது ஒரு சிறந்த துறை."    # Native Tamil
    ]

    print(f"{'LINE CONTENT':<40} | {'DETECTED LANGUAGE'}")
    print("-" * 65)

    for line in mixed_data:
        # Step 1: Detect Native Tamil using Unicode Range
        if re.search(r'[\u0B80-\u0BFF]', line):
            lang = "Tamil (Native Script)"

        else:
            # Step 2: Differentiate English from Tanglish
            # Logic: If it contains common English functional words, it's English
            english_stops = r"\b(the|is|am|are|was|were|will|to|for|of|in|on|at|weather|pleasant|today)\b"
            if re.search(english_stops, line, re.IGNORECASE):
                lang = "English"
            else:
                # Step 3: If Latin script but no English stops, classify as Tanglish
                lang = "Tanglish (Romanized Tamil)"

        print(f"{line:<40} | {lang}")

# Execute the processor
process_mixed_content()
```
