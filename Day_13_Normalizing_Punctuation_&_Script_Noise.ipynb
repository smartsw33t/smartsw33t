```python
import re

def clean_tamil_noise(text):
    # 1. Remove Zero-Width Joiners/Spaces (common 'invisible' noise)
    # \u200B = Zero Width Space, \u200C = ZW Non-Joiner, \u200D = ZW Joiner
    text = re.sub(r'[\u200B-\u200D]', '', text)

    # 2. Normalize Quotes and Dashes
    # Convert various 'fancy' quotes to a standard single quote
    text = re.sub(r'[“”„""]', '"', text)
    text = re.sub(r'[––—]', '-', text)

    # 3. Handle Punctuation attached to Tamil words
    # Language Engineers often add a space before punctuation for better tokenization
    text = re.sub(r'([அ-ஹ])([,\.!\?])', r'\1 \2', text)

    # 4. Collapse multiple spaces into one
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Example of 'dirty' input
dirty_data = "தமிழ்\u200Bமொழி—மிகவும்“   பழமையானது”. "
print(f"Dirty: {dirty_data}")
print(f"Clean: {clean_tamil_noise(dirty_data)}")
```
