```python
import re  # Step 1: Import the built-in Regular Expression library

def regex_day_1():
    # The Tamil text we want to search within
    sample_text = "அன்பு இலார் எல்லாம் தமக்குரியர் அன்பு உடையார் என்பும் உரியர் பிறர்க்கு."

    # The pattern we are looking for: "அன்பு"
    # In Regex, writing the word itself acts as the 'formula' for an exact match.
    pattern = r"அன்பு"

    # re.findall() searches the text and returns a list of every instance it finds.
    matches = re.findall(pattern, sample_text)

    # Output the results
    print(f"Original Text: {sample_text}")
    print(f"Searching for: {pattern}")
    print(f"Number of times found: {len(matches)}")
    print(f"Matches list: {matches}")

# Run the function
regex_day_1()
```
